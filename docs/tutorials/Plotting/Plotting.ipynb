# Histograms and Density Plots

```python
# Pandas and numpy for data manipulation
import pandas as pd
import numpy as np

# Matplotlib and seaborn for plotting
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['figure.dpi'] = 300

import seaborn as sns

from IPython.core.pylabtools import figsize
```

# Data Import 

```python
 # Read in data
ProfitMax = pd.read_csv('D:/Development/GitHub-ProjectV2.0/MaaSSim/docs/tutorials/Plotting/ProfitMax.csv')
PooledRide = pd.read_csv('D:/Development/GitHub-ProjectV2.0/MaaSSim/docs/tutorials/Plotting/PooledRide.csv')
PrivateRide = pd.read_csv('D:/Development/GitHub-ProjectV2.0/MaaSSim/docs/tutorials/Plotting/PrivateRide.csv')

ProfitMax.head(10)
```

# Profit Maximization 

```python
figsize(3, 2)

# Make the histogram using matplotlib, bins must be integet
sns.distplot(ProfitMax['REVENUE'], hist=True, kde=True, color = 'blue',
             hist_kws={'edgecolor':'black'})

plt.title('Revenue Distribution of Individual drivers')
plt.xlabel('Revenue'); plt.ylabel('Drivers');

#ax = sns.displot(df.REVENUE, kde=True, binwidth=1)
```

# Density Plot of All Delays Using Seaborn

```python
figsize(3, 2)

sns.distplot(ProfitMax['REVENUE'], hist=True, kde=True, 
             bins=int(100/5), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 1});

plt.title('Profit Max. Revenue Distribution of Individual drivers'); 
plt.xlabel('Revenue'); plt.ylabel('Drivers');
```

```python
figsize(3, 2)

sns.distplot(PooledRide['REVENUE'], hist=True, kde=True, 
             bins=int(100/5), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 1});

plt.title('PooledRide. Revenue Distribution of Individual drivers'); 
plt.xlabel('Revenue'); plt.ylabel('Drivers');
```

```python
figsize(3, 2)

sns.distplot(PrivateRide['REVENUE'], hist=True, kde=True, 
             bins=int(100/5), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 1});

plt.title('PrivateRide. Revenue Distribution of Individual drivers'); 
plt.xlabel('Revenue'); plt.ylabel('Drivers');
```

# Shaded Density 

```python
figsize(3, 2)

sns.kdeplot(ProfitMax['REVENUE'], shade=True);
sns.kdeplot(PooledRide['REVENUE'], shade=True);
sns.kdeplot(PrivateRide['REVENUE'], shade=True);


plt.title('Revenue Distribution of Individual drivers'); 
plt.xlabel('Revenue'); plt.ylabel('Drivers');
```

```python
# Make a separate list for each airline
x1 = list(ProfitMax['REVENUE'])
x2 = list(PooledRide['REVENUE'])
x3 = list(PrivateRide['REVENUE'])


# Assign colors for each airline and the names
colors = ['#E69F00', '#56B4E9', '#009E73']
names = ['Profit Maximization', 'Pooled Rides', 'Private Rides']


figsize(5, 3)

# Make the histogram using a list of lists
# Normalize the flights and assign colors and names
plt.hist([x1, x2, x3], bins = int(100/10), 
         color = colors, label=names)

# Plot formatting
plt.legend()
plt.xlabel('Revenue')
plt.ylabel('Drivers')
plt.title('Revenue Distributions');

```
