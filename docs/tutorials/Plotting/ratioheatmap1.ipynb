```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.rcParams["figure.figsize"] = [12.50, 3.50]
plt.rcParams["figure.autolayout"] = True


solo = pd.read_csv('D:/Development/GitHub-ProjectV2.0/MaaSSim/docs/tutorials/Results/Simulation/Driver/ServiceRate/Solo-Rides.csv')
profitmax =pd.read_csv('D:/Development/GitHub-ProjectV2.0/MaaSSim/docs/tutorials/Results/Simulation/Driver/ServiceRate/Profit-Maximization.csv')
pooled =pd.read_csv('D:/Development/GitHub-ProjectV2.0/MaaSSim/docs/tutorials/Results/Simulation/Driver/ServiceRate/Pooled-Rides.csv')

fig, (ax1, ax2, ax3) = plt.subplots(ncols=3)
fig.subplots_adjust(wspace=1.01)


s = sns.heatmap(profitmax, annot=True,  fmt=".2f", cmap='viridis', ax=ax1, vmin=0,vmax=100)
s.set(xlabel='number of Travellers', ylabel='number of Drivers', title = 'Service Rate for Profit Maximization')

s = sns.heatmap(pooled, annot=True, fmt=".2f", cmap='viridis', ax=ax2, vmin=0,vmax=100)
s.set(xlabel='number of Travellers', ylabel='number of Drivers', title = 'Service Rate for Pooled Rides')

s = sns.heatmap(solo, annot=True, fmt=".2f", cmap='viridis', ax=ax3, vmin=0,vmax=100)
s.set(xlabel='number of Travellers', ylabel='number of Drivers', title = 'Service Rate for Solo Rides')

txt="I need the caption to be present a little below X-axis"
#ax3.yaxis.tick_right()

fig.subplots_adjust(wspace=0.001)
plt.show()
fig.savefig('ServiceRate.png', dpi=600)
```
