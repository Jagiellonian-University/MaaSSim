# KPI-1: Profit 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.rcParams["figure.figsize"] = [12.50, 3.50]
plt.rcParams["figure.autolayout"] = True

solo = pd.read_csv('D:/Development/GitHub-ProjectV2.0/MaaSSim/docs/tutorials/Results/Simulation/Driver/KPI-1-Profit/KPI2_Solo-Rides.csv')
solo=solo.drop('0',axis='columns')

profitmax =pd.read_csv('D:/Development/GitHub-ProjectV2.0/MaaSSim/docs/tutorials/Results/Simulation/Driver/KPI-1-Profit/KPI1_Profit-Maximization.csv')
profitmax=profitmax.drop('0',axis='columns')

pooled =pd.read_csv('D:/Development/GitHub-ProjectV2.0/MaaSSim/docs/tutorials/Results/Simulation/Driver/KPI-1-Profit/KPI3_Pooled-Rides.csv')
pooled=pooled.drop('0',axis='columns')
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3)
fig.subplots_adjust(wspace=1.01)


s = sns.heatmap(profitmax, annot=True,  fmt=".2f", cmap='viridis', ax=ax1, vmin=0, vmax=200)
s.set(xlabel='number of Travellers', ylabel='number of Drivers', title = 'Profit for Profit Maximization')
s.invert_yaxis()
s = sns.heatmap(pooled, annot=True, fmt=".2f", cmap='viridis', ax=ax2, vmin=0, vmax=220)
s.set(xlabel='number of Travellers', ylabel='number of Drivers', title = 'Profit for Pooled Rides')
s.invert_yaxis()
s = sns.heatmap(solo, annot=True, fmt=".2f", cmap='viridis', ax=ax3,  vmin=0, vmax=200)
s.set(xlabel='number of Travellers', ylabel='number of Drivers', title = 'Profit for Solo Rides')
s.invert_yaxis()
txt="I need the caption to be present a little below X-axis"
#ax3.yaxis.tick_right()


fig.subplots_adjust(wspace=0.001)
plt.show()
#fig.savefig('ServiceRate.png', dpi=600)
```
