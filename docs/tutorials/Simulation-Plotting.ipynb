```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define a function to create a heatmap plot
def create_heatmap_subplot(data_file, title, subplot_num):
    # Load the data
    occupancy_data = pd.read_csv(data_file, index_col=0)

    # Define number of drivers and travelers
    drivers = np.arange(10, 51, 10)
    travelers = np.arange(100, 501, 100)

    # Create a DataFrame for occupancy
    df_occupancy = pd.DataFrame(occupancy_data.values, index=travelers, columns=drivers)

    # Create the subplot
    plt.subplot(1, 3, subplot_num)

    # Create the heatmap using Seaborn with a different colormap and annotations
    heatmap = sns.heatmap(df_occupancy, annot=True, fmt='.3f', vmax=1, cmap='viridis', linewidths=0.5, cbar_kws={'label': 'Occupancy'})

    # Reverse the y-axis
    plt.gca().invert_yaxis()

    # Set y-axis labels
    plt.xlabel('Number of Drivers')
    plt.ylabel('Number of Travelers')
    plt.title(title)

# Create a figure with three subplots in a single row
plt.figure(figsize=(20, 6))


# Plot 2
create_heatmap_subplot('D:/Development/GitHub-ProjectV2.0/MaaSSim/docs/tutorials/Results/Simulation/Driver/KPI-8-Occupancy/Rep/Occupancy/Solo_ride_Occupancy.csv', 'Occupancy with Solo Rides', 1)



# Plot 1
create_heatmap_subplot('D:/Development/GitHub-ProjectV2.0/MaaSSim/docs/tutorials/Results/Simulation/Driver/KPI-8-Occupancy/Rep/Occupancy/Profit_Max_Occupancy.csv', 'Occupancy with Profit Max', 2)


# Plot 3
create_heatmap_subplot('D:/Development/GitHub-ProjectV2.0/MaaSSim/docs/tutorials/Results/Simulation/Driver/KPI-8-Occupancy/Rep/Occupancy/Pooled_ride_Occupancy.csv', 'Occupancy with Pooled Rides', 3)

# Adjust subplot layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# -*- coding: utf-8 -*-
"""
Created on Wed Aug 9 12:54:58 2023
@author: Usman Akhtar
"""
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('D:\Development\GitHub-ProjectV2.0\MaaSSim\docs\Experiments\data.csv')

data.fillna(0, inplace=True)

time_values = data['Time']
service_rate1_values = data['service rate1']
service_rate2_values = data['service rate2']
service_rate3_values = data['service rate3']

# Combine data for Seaborn
seaborn_data = pd.melt(data, id_vars='Time', value_vars=['service rate1', 'service rate2', 'service rate3'],
                       value_name='Service Rate %', var_name='Service')

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Define a custom palette with colors for the bars
custom_palette = ['b', 'g', 'r']
sns.set_palette(custom_palette)

# Create a bar plot using Seaborn with hatching patterns
ax = sns.barplot(data=seaborn_data, x='Time', y='Service Rate %', hue='Service', alpha=0.8)

# Apply hatching patterns to each bar
hatches = ['//', '\\\\', 'x']
for i, patch in enumerate(ax.patches):
    hatch = hatches[i % len(hatches)]
    patch.set_hatch(hatch)

# Adjust other formatting
ax.set_xlabel('Time', fontsize=12)
ax.set_ylabel('Service Rate %', fontsize=12)
ax.set_title('Service Rates of Different KPIs', fontsize=14)
ax.tick_params(axis='both', which='major', labelsize=10)
ax.legend(title='Service', title_fontsize=10, fontsize=10)

plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```
